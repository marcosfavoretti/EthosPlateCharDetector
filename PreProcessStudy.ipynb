```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
image = cv2.imread('4study/study0count.png')

# Converter para escala de cinza
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Aplicar a equalizao do histograma
equalized = cv2.equalizeHist(gray)

# Aplicar a filtragem gaussiana para suavizar a imagem
blurred = cv2.GaussianBlur(equalized, (5, 5), 0)

# Aplicar a deteo de bordas com o operador Laplaciano
edges = cv2.Laplacian(blurred, cv2.CV_8U, ksize=9)

# Aplicar um limiar para binarizar a imagem
_, thresh = cv2.threshold(edges, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Encontrar contornos na imagem binarizada
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Criar uma mscara binria do mesmo tamanho da imagem original
mask = np.zeros_like(gray, dtype=np.uint8)

# Desenhar os contornos na mscara binria
cv2.drawContours(mask, contours, -1, (255), thickness=cv2.FILLED)

# Aplicar a mscara binria na imagem original para remover os focos luminosos
result = cv2.bitwise_and(image, image, mask=mask)

# Mostrar o resultado
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
